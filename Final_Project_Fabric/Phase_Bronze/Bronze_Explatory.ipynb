```python
"""
FIRST: UNDERSTAND YOUR Bronze_Open_Air TABLE STRUCTURE
"""

print("="*80)
print("üîç ANALYZING YOUR Bronze_Open_Air TABLE")
print("="*80)

from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()

# 1. Load your table
df = spark.read.table("dbo.Bronze_Open_Air")

# 2. Show basic info
print(f"üìä TABLE OVERVIEW:")
print(f"   Total rows: {df.count():,}")
print(f"   Total columns: {len(df.columns)}")
print(f"   Columns: {', '.join(df.columns)}")

# 3. Show schema (data types)
print(f"\nüìã SCHEMA (Data Types):")
df.printSchema()

# 4. Show sample data with all columns
print(f"\nüîé SAMPLE DATA (First 5 rows, all columns):")
# Create a truncated view of all columns
sample = df.limit(5)
for row in sample.collect():
    print(f"\nRow:")
    for col_name in df.columns:
        value = row[col_name]
        # Truncate long values
        if value and len(str(value)) > 50:
            value = str(value)[:50] + "..."
        print(f"  {col_name:25}: {value}")

# 5. Check unique values in key columns
print(f"\nüéØ KEY METRICS:")
print(f"   Unique parameters: {df.select('param_name').distinct().count()}")
print(f"   Unique cities: {df.select('city').distinct().count()}")
print(f"   Date range:")
min_date = df.select(min('measurement_date')).collect()[0][0]
max_date = df.select(max('measurement_date')).collect()[0][0]
print(f"     From: {min_date}")
print(f"     To:   {max_date}")

# 6. Check data quality issues
print(f"\n‚ö†Ô∏è  DATA QUALITY ISSUES:")
print(f"   Rows with NULL local_datetime: {df.filter(col('local_datetime').isNull()).count():,}")
print(f"   Rows with future dates: {df.filter(col('measurement_date') > '2024-12-31').count():,}")
print(f"   Negative values: {df.filter(col('value') < 0).count():,}")

# 7. Parameter distribution
print(f"\nüìà PARAMETER DISTRIBUTION:")
param_counts = df.groupBy('param_name').count().orderBy('count', ascending=False)
for row in param_counts.collect():
    print(f"   {row['param_name']:10}: {row['count']:>6} rows")

print("\n" + "="*80)
print("üìã PLEASE SHARE THIS OUTPUT SO I CAN CREATE THE PERFECT SILVER TRANSFORMATION")
print("="*80)
```
